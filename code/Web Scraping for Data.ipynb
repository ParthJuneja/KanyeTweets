# Web Scrapping for tweets

```python
# imports
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
#global array for storing tweets
Kanye_Tweets = []
kanye_acc = 169686021
```

```python
#function for storing tweets from page
def fetch_tweets_from_page(account,page_no,dell=''):
    url = ('https://polititweet.org/tweets?page=%d&deleted=%s&account=%d' %(page_no,dell,account))
#     https://polititweet.org/tweets?page=1&deleted=True&account=169686021 for deleted

    # Send an HTTP GET request to the URL
    response = requests.get(url)

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content of the page using BeautifulSoup
        soup = BeautifulSoup(response.text, 'html.parser')

        # Find the elements containing the text you want to scrape
        tweet_page = soup.find_all('p',class_="small-top-margin")

        #list for all tweets
        all_tweets_from_this_page = []

        for twt in tweet_page:
            twt_text = (str("".join(twt.get_text())))
            all_tweets_from_this_page.append(twt_text.rstrip('\n‚Äî\xa0PolitiTweet.org\n').lstrip('\n'))

        return all_tweets_from_this_page

    else:
        print('Failed to retrieve the webpage. Status code:', response.status_code)
        return
```

```python
#116 archived 17 deleted
for i in range(1,117):
    Kanye_Tweets.extend(fetch_tweets_from_page(kanye_acc,i))
    print(i)
```

```python
#116 archived 17 deleted
for i in range(1,18):
    Kanye_Tweets.extend(fetch_tweets_from_page(kanye_acc,i,'True'))
    print(i)
```
